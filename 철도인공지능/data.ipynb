```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV   
csv_file_path = 'data_c30.csv'

# CSV  
data = pd.read_csv(csv_file_path)

#    
selected_columns = ['Distance', 'YL_M1_B1_W1', 'YR_M1_B1_W1', 'YL_M1_B1_W2', 'YR_M1_B1_W2']
selected_data = data[selected_columns]

#   
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
axes = axes.flatten()

#    
for i, column in enumerate(selected_columns[1:]):
    ax = axes[i]
    ax.plot(selected_data['Distance'], selected_data[column])
    ax.set_xlabel('Distance')
    ax.set_ylabel('Values')
    ax.set_title(column)
    ax.grid()

#     
plt.tight_layout()
plt.show()
```
